```python
#1.7 tính chỉ số BMI
import numpy as np

names = np.array(['Ann','Joe','Mark'])
heights = np.array([1.5, 1.78, 1.6])
weights = np.array([65, 46, 59])

bmi = weights/heights **2 

for i, cs in enumerate(bmi):
    if cs < 18.5:
        print(names[i], "Underweight BMI:", np.around(cs,2))
    else:
        if cs>25:
            print(names[i], "Overweight BMI:", np.around(cs,2))
        else:
            print(names[i], "Normal weight BMI:", np.around(cs,2))
```

```python
#1.9 # vẽ biểu đồ thể hiện tỉ lên nam nữ
import pandas as pd
import seaborn as sns #tạo biểu đồ và thống kê dữ liệu
import matplotlib.pyplot as plt

data = pd.read_csv('x1.csv')

plt.figure(figsize=(8, 6))  # Đặt kích thước của biểu đồ
#tạo biểu đồ cột , errorbar=None không hiển thị khoảng tin cậy
sns.barplot(x="gender", y="license", hue="group", data=data, errorbar=None)

plt.title("Proportion with Driving License")
plt.xlabel("Gender") #nhãn trục ox
plt.ylabel("Proportion") #nhãn trục oy
plt.xticks(ticks=[0, 1], labels=["Men", "Women"]) #đặt tên cho các nhóm trên trục ox

plt.legend(title="Group") #hiển thị chú thích
plt.show()
```

```python
# 1.10 #biểu đồ tỉ lệ sống sót của các hành khách giữa các nhóm

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

titanic = sns.load_dataset("titanic") #đọc như liệu của titatic có sẵn từ seaborn vào dataframe
#tạo biểu đồ cột
#trục ox, oy, chia biểu đồ thành các cột, không hiển thị khoảng tin cậy, tỉ lệ biểu đồ
g = sns.catplot(x="who", y="survived", col="class", 
 data=titanic, kind="bar", errorbar=None, aspect=1)
g.set_axis_labels("", "Survival Rate") #đặt nhãn cho trục x, y
g.set_xticklabels(["Men", "Women", "Children"]) #đặt nhãn cho các nhóm trên ox
g.set_titles("{col_name} {col_var}") #đặt tiêu đề cho các cột
#---show plot---
plt.show()
```

```python
#12
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Khai báo dữ liệu đầu vào
X = np.array([50, 60, 65, 70, 75, 80, 85]).reshape(-1, 1)  # (mảng 2 chiều)Reshape để đảm bảo kích thước phù hợp
y = np.array([2.5, 3, 3.5, 3.8, 4, 4.5, 5])

# Khởi tạo mô hình hồi quy tuyến tính giúp dự đoán giá nhà trên diện tích của nhà
model = LinearRegression()

# Huấn luyện mô hình trên dữ liệu
# yêu câu đầu vào là 1 mảng 2 chiều cho biến độc lập X và 1 mảng 1 chiều cho biến phụ thuộc y
model.fit(X, y)

# Dự đoán giá nhà cho các diện tích mới
X_new = np.array([55, 68, 76, 90]).reshape(-1, 1)
y_new = model.predict(X_new) # dự đoán giá nhà cho diện tích mới

# Vẽ biểu đồ phân tán dữ liệu và đường hồi quy
plt.scatter(X, y)
plt.plot(X_new, y_new, color='red')
plt.xlabel('Diện tích (m2)')
plt.ylabel('Giá nhà (tỷ)')
plt.show()

for i in range(len(X_new)):
    print(f"Diện tích {X_new[i][0]}m2: Dự đoán giá nhà = {round(y_new[i], 2)} tỷ")

```

```python
#13 dự đoán cân nặng khi cho chiều cao

%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# represents the heights of a group of people in meters
heights = [[1.6], [1.65], [1.7], [1.73], [1.8]]
# represents the weights of a group of people in kgs
weights = [[60], [65], [72.3], [75], [80]]
plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters')
plt.ylabel('Weights in kilograms')
plt.plot(heights, weights, 'k.')
# axis range for x and y
plt.axis([1.5, 1.85, 50, 90])
plt.grid(True)

# khởi tạo hồi quy tuyến tính
model = LinearRegression()
#huấn luyện mô hình trên dữ liệu
model.fit(X=heights, y=weights)
# dự đoán câm nặng với chiều cao mới
weight = model.predict([[1.9]])[0][0]
print(round(weight,2)) 
```

```python
#14
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

#Chuẩn bị dữ liệu bằng cách chia dữ liệu thành hai phần, một phần chứa 
#các đặc trưng (features) và phần còn lại chứa biến mục tiêu (target).
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])

#raw_df.values[1::2, :2] chứa các giá trị của biến mục tiêu (target), trong trường hợp này là 'MEDV'.
target = raw_df.values[1::2, 2]
print(data)
```
